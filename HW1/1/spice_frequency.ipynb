```python
import pandas as pd
from matplotlib import pyplot as plt

df= pd.read_fwf('spice.din.txt')
```

```python
split_columns=df['type_address'].str.split(" ",n=1,expand=True)
split_columns=split_columns.rename(columns={0:'type',1:'address'})
type=split_columns['type']
```

```python
plt.hist(type, edgecolor='black')
plt.xlabel("CPU operation 0: Read, 1:Write, 2: Instruction fetch ")
plt.ylabel("Frequency")
plt.title('Frequencies of read write and instruction fetch')
```
